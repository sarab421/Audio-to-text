```python
# Import required libraries
import speech_recognition as sr
from pydub import AudioSegment
# Function to convert audio to text using Google Web Speech API
def transcribe_audio(file_path):
    print(file_path);
    """
    Transcribe an audio file to text using Google Web Speech API.

    Parameters:
        file_path (str): Path to the audio file.

    Returns:
        str: Transcribed text.
    """
    # Initialize recognizer
    recognizer = sr.Recognizer()

    # Load the audio file
    with sr.AudioFile(file_path) as source:
        print("Loading audio...")
        audio_data = recognizer.record(source)  # Capture entire audio file

    # Perform transcription
    try:
        print("Transcribing...")
        transcription = recognizer.recognize_google(audio_data)
        return transcription
    except sr.UnknownValueError:
        return "Google Web Speech API could not understand the audio."
    except sr.RequestError as e:
        return f"Could not request results from Google Web Speech API; {e}"

# Example Usage
file_path = "/content/NASTP Kharian - National Aerospace Science & Technology Park.wav"  # Replace with the path to your audio file
result = transcribe_audio(file_path)
print("Transcribed Text:", result)
```

```python
from TTS.api import TTS
from datetime import datetime

def clone_voice_and_mimic_prosody(
    text: str,
    reference_audio: str,
    output_file: str,
    model_name: str = "tts_models/multilingual/multi-dataset/your_tts",
    language: str = "en"
):
    """
    Clones the voice and mimics the rhythm and accent based on the reference audio.

    Parameters:
        text (str): The input text to synthesize.
        reference_audio (str): Path to the reference audio file for cloning.
        output_file (str): Path where the output audio will be saved.
        model_name (str): The pre-trained TTS model to use (default: multilingual).
        language (str): Language code for the model (default: "en").

    Returns:
        None
    """
    print(f"Initializing TTS model: {model_name}")
    tts = TTS(model_name=model_name)

    start_time = datetime.now()
    print(f"Start time: {start_time.strftime('%H:%M:%S')}")

    # Generate speech with cloned voice, rhythm, and accent
    print(f"Cloning voice and mimicking prosody for text: '{text}'")
    tts.tts_to_file(
        text=text,
        speaker_wav=reference_audio,
        language=language,
        file_path=output_file,
    )

    end_time = datetime.now()
    print(f"End time: {end_time.strftime('%H:%M:%S')}")
    print(f"Speech synthesized and saved to: {output_file}")


if __name__ == "__main__":
    # User inputs
    input_text = "My phone is on the table. Where is my phone?"
    reference_audio_path = "/content/NASTP Kharian - National Aerospace Science & Technology Park.wav"
    output_audio_path = "/content/output_cloned_voice.wav"

    # Call the function
    clone_voice_and_mimic_prosody(
        text=input_text,
        reference_audio=reference_audio_path,
        output_file=output_audio_path,
    )
```
